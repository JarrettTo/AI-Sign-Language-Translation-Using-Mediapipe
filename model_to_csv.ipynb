```python
import numpy as np
import pandas as pd
import os
```

```python
# variables to be changed
label = "thanks"
seq_num = 30
frame_num = 30

columns = ['Label', 'Coordinates']

# Create an empty list to store DataFrames
data_frames = []

for i in range(1, seq_num + 1):
    sequence = []
    for j in range(frame_num):
        # Construct the file path using label, i, and j
        npy_file_path = f'MP_Data/{label}/{i}/{j}.npy'
        
        # Load data from the current .npy file
        data = np.load(npy_file_path)
        
        # Append data to the sequence
        sequence.append(data)
        
    # Append the DataFrame to the list
    data_frames.append((label, sequence))

# Now, data_frames contains DataFrames from all the specified .npy files
final_df = pd.DataFrame(data_frames, columns=columns)

directory_path = "MP_Data_csv"
if not os.path.exists(directory_path):
    os.makedirs(directory_path)
    
csv_file_path = f'{directory_path}/{label}.csv'
final_df.to_csv(csv_file_path, index=False, mode='w')
```
